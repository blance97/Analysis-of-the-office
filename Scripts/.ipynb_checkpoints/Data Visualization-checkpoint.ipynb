```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./Data/the-office-lines - scripts.csv')
df.head(3)
```

## Words Spoken By Character(Season 1)

```python
Michael_Count = df.query('season == 1 & speaker =="Michael"')['line_text'].str.count("\\s")+1
Jim_Count = df.query('season == 1 & speaker =="Jim"')['line_text'].str.count("\\s")+1
Pam_Count = df.query('season == 1 & speaker =="Pam"')['line_text'].str.count("\\s")+1
Dwight_Count = df.query('season == 1 & speaker =="Dwight"')['line_text'].str.count("\\s")+1
Ryan_Count = df.query('season == 1 & speaker =="Ryan"')['line_text'].str.count("\\s")+1
```

```python
d = {'Michael_Count': Michael_Count.sum(), 'Jim_Count': Jim_Count.sum(), 'Pam_Count':Pam_Count.sum(),'Dwight_Count':Dwight_Count.sum(),'Ryan_Count' }
WordsSpokenS1 = pd.Series(d, name='WordsSpokenS1')
plt.title('Season 1(Spoke most words)')
WordsSpokenS1.plot.bar()
```

## Spoke Most Lines by Season

```python
season1 = df.query('season == 1')
season2 = df.query('season == 2')
season3 = df.query('season == 3')
season4 = df.query('season == 4')
season5 = df.query('season == 5')
season6 = df.query('season == 6')
season7 = df.query('season == 7')
season8 = df.query('season == 8')
season9 = df.query('season == 9')
```

```python
val = season1.groupby('speaker').count().sort_values(by=['id'],ascending=False)
plt.title('Season 1(Spoke most lines)')
val["id"].plot.bar()
```

```python
val = season2.groupby('speaker').count().sort_values(by=['id'],ascending=False)
plt.title('Season 2(Spoke most lines)')
val["id"].plot.bar()
```

```python
val = season3.groupby('speaker').count().sort_values(by=['id'],ascending=False)
plt.title('Season 3(Spoke most lines)')
val["id"].plot.bar()
```
