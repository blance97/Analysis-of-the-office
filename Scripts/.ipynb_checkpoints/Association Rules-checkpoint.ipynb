```python
import pandas as pd
from mlxtend.preprocessing import OnehotTransactions
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../Data/the-office-lines - scripts.csv')
df = df.query('deleted == False')
df = df.drop(['id','line_text','deleted'],axis=1)
df.head(20)
```

```python
s1 = df.query('season == 1')
num_eps = len(s1['episode'].unique())

g = s1.groupby(['episode','scene'])

convos = []
eps = s1.groupby(['episode'])
for i in range(num_eps):
    num_scenes = eps.get_group(i+1)['scene'].unique()
    for j in num_scenes:
        speakers = g.get_group((i+1,j))['speaker'].unique()
        convos.append(speakers)
```

```python
oht = OnehotTransactions()
oht_ary = oht.fit(convos).transform(convos)
s1_coded = pd.DataFrame(oht_ary, columns=oht.columns_)
s1_coded.head()
```

```python
# season 1 frequent itemsets
frequent_itemsets = apriori(s1_coded, min_support=0.1, use_colnames=True)
frequent_itemsets.round(3)
```

```python
# season 1 association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules.round(3)
```
