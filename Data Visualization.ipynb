## Words Spoken By Character(Season 1)

```python
def plotSpokeWords(WordsSpoken, season):
    plt.title('Season ' + season +' (Most words)')
    ax = WordsSpoken.plot(kind='bar', figsize=(10,7), fontsize=13);
    totals = []
    # find the values and append to list
    for i in ax.patches:
        totals.append(i.get_height())
    total = sum(totals)

    counter = 0;

    for i in ax.patches:
        # get_x pulls left or right; get_height pushes up or down

        if (counter < 20):
            ax.text(i.get_x()-.03, i.get_height()+.5, \
                str(round((i.get_height()), 2)), fontsize=15,
                    color='dimgrey')
        counter+=1

Michael_Count = season1.query('speaker =="Michael"')['line_text'].str.count("\\s")+1
Jim_Count = season1.query('speaker =="Jim"')['line_text'].str.count("\\s")+1
Pam_Count = season1.query('speaker =="Pam"')['line_text'].str.count("\\s")+1
Dwight_Count = season1.query('speaker =="Dwight"')['line_text'].str.count("\\s")+1
Ryan_Count = season1.query('speaker =="Ryan"')['line_text'].str.count("\\s")+1

d = {'Michael Scott': Michael_Count.sum(), 'Jim Halpert': Jim_Count.sum(), 'Pam Beesly':Pam_Count.sum(),'Dwight Schrute':Dwight_Count.sum(),'Ryan Howard':Ryan_Count.sum() }
WordsSpoken = pd.Series(d, name='WordsSpoken')
plotSpokeWords(WordsSpoken, "1")
```

## Spoke Most Lines by Season

```python
def plotSpokeLines(df, season):
    val = df.groupby('speaker').count().sort_values(by=['id'],ascending=False)
    plt.title('Season' + season +'(Spoke most lines)')
    # val["id"].plot.bar()

    ax = val["id"].nlargest(20).plot(kind='bar', figsize=(10,7),
                                            color="coral", fontsize=13);
    # ax.set_title("Where were the battles fought?", fontsize=18)
    # ax.set_ylabel("Number of Battles", fontsize=18);
    totals = []
    # find the values and append to list
    for i in ax.patches:
        totals.append(i.get_height())
    total = sum(totals)

    counter = 0;

    for i in ax.patches:
        # get_x pulls left or right; get_height pushes up or down

        if (counter < 20):
            ax.text(i.get_x()-.03, i.get_height()+.5, \
                str(round((i.get_height()), 2)), fontsize=15,
                    color='dimgrey')
        counter+=1
plotSpokeLines(season1, "1")
```
